Bài tập 1: Tạo mảng và thao tác cơ bản

```python
import numpy as np

# 1. Tạo một mảng NumPy với các giá trị từ 1 đến 20
array_1 = np.arange(1, 21)
print("Mảng từ 1 đến 20:", array_1)

# 2. Tìm tổng, giá trị lớn nhất, nhỏ nhất và trung bình của mảng
print("Tổng:", np.sum(array_1))
print("Giá trị lớn nhất:", np.max(array_1))
print("Giá trị nhỏ nhất:", np.min(array_1))
print("Trung bình:", np.mean(array_1))

# 3. Tạo một mảng 2D (3x5) chứa các số ngẫu nhiên từ 0 đến 100
array_2d = np.random.randint(0, 101, size=(3, 5))
print("Mảng 2D:", array_2d)

# 4. Lấy hàng thứ 2 và cột thứ 3 của mảng 2D
print("Hàng thứ 2:", array_2d[1])
print("Cột thứ 3:", array_2d[:, 2])
```

Bài tập 2: Các thao tác nâng cao


```python
import numpy as np

# 1. Tạo một mảng NumPy chứa 20 giá trị ngẫu nhiên từ 0 đến 1
random_array = np.random.rand(20)
print("Mảng ngẫu nhiên từ 0 đến 1:", random_array)

# 2. Chuẩn hóa mảng này (đưa các giá trị về khoảng [0, 1])
# Giả sử mảng đã có sẵn các giá trị trong khoảng [0, 1], chúng ta không cần phải làm gì thêm.
# Tuy nhiên, nếu mảng có giá trị ngoài khoảng [0, 1], bạn có thể chuẩn hóa như sau:
normalized_array = (random_array - np.min(random_array)) / (np.max(random_array) - np.min(random_array))
print("Mảng sau khi chuẩn hóa:", normalized_array)
# 3. Tính tích vô hướng (dot product) của hai mảng 1D
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
dot_product = np.dot(a, b)
print("Tích vô hướng của a và b:", dot_product)
# 4. Tạo một ma trận 5x5 và tính định thức, nghịch đảo
matrix = np.random.randint(1, 10, size=(5, 5))  # Ma trận 5x5 với các số ngẫu nhiên từ 1 đến 9
print("Ma trận:", matrix)
determinant = np.linalg.det(matrix)
print("Định thức của ma trận:", determinant)
if determinant != 0:
    inverse_matrix = np.linalg.inv(matrix)
    print("Ma trận nghịch đảo:", inverse_matrix)
else:
    print("Ma trận không khả nghịch (định thức = 0).")
```

Bài tập 3: Làm quen với DataFrame

```python
import pandas as pd

# 1. Tạo DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [23, 25, 22, 24, 21],
    'Score': [85, 90, 78, 92, 88]
}
df = pd.DataFrame(data)
print("DataFrame:\n", df)
# 2. Tính giá trị trung bình của cột "Score"
average_score = df['Score'].mean()
print("Điểm trung bình:", average_score)
# 3. Lọc các hàng có "Score" lớn hơn 85
filtered_df = df[df['Score'] > 85]
print("Các hàng có Score > 85:\n", filtered_df)
```

Bài tập 4: Đọc và phân tích dữ liệu từ file

```python
import pandas as pd
# 1. Đọc file CSV (thay đổi đường dẫn đến file Iris.csv trên máy của bạn)
iris_df = pd.read_csv('Iris.csv')  # Đảm bảo rằng bạn đã tải file Iris.csv và cung cấp đường dẫn đúng
print("Thông tin tổng quan về dữ liệu:\n", iris_df.info())
print("Mô tả dữ liệu:\n", iris_df.describe())
# 2. Tính toán cơ bản cho cột 'sepal_length'
print("Trung bình sepal_length:", iris_df['sepal_length'].mean())
print("Giá trị lớn nhất sepal_length:", iris_df['sepal_length'].max())
print("Giá trị nhỏ nhất sepal_length:", iris_df['sepal_length'].min())
```

Phần 3: Làm sạch dữ liệu

```python
import pandas as pd

# 1. Tạo DataFrame chứa dữ liệu thiếu
data_with_missing = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [23, None, 25, 24, 22],
    'City': ['New York', 'Boston', None, 'Chicago', 'Boston'],
    'Salary': [60000, 52000, None, 58000, None]
}
df_missing = pd.DataFrame(data_with_missing)
print("Dữ liệu ban đầu:\n", df_missing)
# 2. Điền giá trị thiếu trong cột Age bằng giá trị trung bình
mean_age = df_missing['Age'].mean()
df_missing['Age'].fillna(mean_age, inplace=True)
print("\nSau khi điền thiếu ở Age:\n", df_missing)
# 3. Xóa các hàng có nhiều hơn 1 giá trị thiếu
df_missing.dropna(thresh=len(df_missing.columns)-1, inplace=True)  # Xóa hàng có nhiều hơn 1 giá trị thiếu
print("\nSau khi xóa hàng thiếu nhiều hơn 1 cột:\n", df_missing)
# 4. Điền giá trị thiếu trong cột Salary bằng 50000
df_missing['Salary'].fillna(50000, inplace=True)
print("\nSau khi điền thiếu ở Salary:\n", df_missing)
```

Phần 4: Trực quan hóa dữ liệu với Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
# 1. Biểu đồ đường hàm số y = x^2 trên khoảng [-10, 10]
x = range(-10, 11)  # Tạo các giá trị x từ -10 đến 10
y = [i**2 for i in x]  # Hàm số y = x^2
plt.plot(x, y, label="y = x^2")
plt.title("Biểu đồ đường hàm số y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
# 2. Biểu đồ cột điểm số từ Bài tập 3
names = ['Alice', 'Bob', 'Charlie', 'David', 'Eva']
scores = [85, 90, 78, 92, 88]
plt.bar(names, scores, color='skyblue')
plt.title("Biểu đồ cột điểm số")
plt.xlabel("Tên sinh viên")
plt.ylabel("Điểm số")
plt.show()
# 3. Biểu đồ tròn thể hiện phần trăm mỗi loại hoa trong tập dữ liệu Iris
# Giả sử ta có một phần dữ liệu Iris:
iris_data = {
    'species': ['setosa', 'versicolor', 'virginica', 'setosa', 'versicolor', 'setosa', 'virginica', 'virginica', 'versicolor', 'setosa']
}
df_iris = pd.DataFrame(iris_data)
# Đếm số lượng mỗi loài hoa
species_counts = df_iris['species'].value_counts()
# Vẽ biểu đồ tròn
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Biểu đồ tròn tỷ lệ mỗi loài hoa trong tập dữ liệu Iris")
plt.show()
```

Bài tập 7: Biểu đồ nâng cao

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Giả sử chúng ta đã có dữ liệu Iris
# Đây là dữ liệu Iris đơn giản để minh họa
iris_data = {
    'sepal_length': [5.1, 4.9, 4.7, 4.6, 5.0, 5.4, 4.6, 5.0, 5.4, 4.8],
    'sepal_width': [3.5, 3.0, 3.2, 3.1, 3.6, 3.9, 3.4, 3.4, 3.7, 3.4],
    'species': ['setosa', 'setosa', 'setosa', 'setosa', 'setosa', 'versicolor', 'versicolor', 'versicolor', 'versicolor', 'versicolor']
}

df_iris = pd.DataFrame(iris_data)

# 1. Biểu đồ phân tán (scatter plot) với màu sắc theo loại hoa
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df_iris, palette='Set1')

# 2. Thêm tiêu đề, nhãn trục và chú thích cho biểu đồ
plt.title("Biểu đồ phân tán giữa Sepal Length và Sepal Width", fontsize=14)
plt.xlabel("Sepal Length (cm)", fontsize=12)
plt.ylabel("Sepal Width (cm)", fontsize=12)
plt.legend(title='Loại Hoa')

plt.show()
```
