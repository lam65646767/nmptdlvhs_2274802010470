```python
import numpy as np

# Tạo mảng
arr = np.array([1, 3, 5, 7, 9])

# 1. Các phép toán cộng, trừ, nhân, chia
arr_add = arr + 2  # Cộng 2 vào từng phần tử
arr_sub = arr - 1  # Trừ 1 từ từng phần tử
arr_mul = arr * 2  # Nhân từng phần tử với 2
arr_div = arr / 2  # Chia từng phần tử cho 2

print("1. Cộng, trừ, nhân, chia:")
print("Cộng 2:", arr_add)
print("Trừ 1:", arr_sub)
print("Nhân 2:", arr_mul)
print("Chia 2:", arr_div)
print()

# 2. Tính tổng, trung bình, lớn nhất, nhỏ nhất
sum_arr = np.sum(arr)           # Tổng các phần tử
mean_arr = np.mean(arr)         # Giá trị trung bình
max_arr = np.max(arr)           # Phần tử lớn nhất
min_arr = np.min(arr)           # Phần tử nhỏ nhất

print("2. Tính tổng, trung bình, lớn nhất, nhỏ nhất:")
print("Tổng:", sum_arr)
print("Trung bình:", mean_arr)
print("Lớn nhất:", max_arr)
print("Nhỏ nhất:", min_arr)
print()

# 3. Lấy phần tử bất kỳ trong mảng
index = 2  # Chọn phần tử ở vị trí thứ 3 (chỉ số bắt đầu từ 0)
random_element = arr[index]

print("3. Lấy phần tử bất kỳ:")
print(f"Phần tử tại vị trí {index}:", random_element)
```

Pands

```python
import pandas as pd 
# 1. Tạo DataFrame
data = {
    "Tên": ["Anh", "Đạt", "Trâm", "Linh"],
    "Tuổi": [21, 25, 27, 18],
    "Điểm": [55, 85, 100, 85]
}
df = pd.DataFrame(data)
print("DataFrame:")
print(df)
print()
# 2. Lọc các hàng có điểm >= 80
filtered_df = df[df["Điểm"] >= 80]
print("Các học sinh có điểm >= 80:")
print(filtered_df)
print()
# 3. Lọc các học sinh có điểm >= điểm trung bình
average_score = df["Điểm"].mean()  # Tính điểm trung bình
above_average_df = df[df["Điểm"] >= average_score]  # Lọc các hàng có Điểm >= điểm trung bình
print(f"Điểm trung bình: {average_score}")
print("Các học sinh có điểm >= điểm trung bình:")
print(above_average_df)

```

Matplotlib


```python
import matplotlib.pyplot as plt
# Dữ liệu (đã sửa)
x = [1, 3, 5, 7]
y = [9, 87, 2, 4]
# Vẽ biểu đồ đường
plt.plot(x, y, color='blue', marker='o', label='Biểu đồ đường')  # Đường màu xanh dương, marker là hình tròn
plt.title("Biểu đồ đường")  # Tiêu đề
plt.xlabel("Trục X")  # Nhãn trục X
plt.ylabel("Trục Y")  # Nhãn trục Y
plt.legend()  # Hiển thị chú thích
plt.grid(True)  # Hiển thị lưới
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
# Dữ liệu
ten = ["Anh", "Đạt", "Trâm", "Linh"]
tuoi = [21, 25, 27, 18]
diem = [55, 85, 100, 85]
# Thiết lập vị trí các cột
x = np.arange(len(ten))  # Vị trí các cột
# Kích thước hình vẽ
plt.figure(figsize=(10, 6))
# Vẽ các cột
plt.bar(x - 0.2, tuoi, width=0.4, color='skyblue', label='Tuổi')
plt.bar(x + 0.2, diem, width=0.4, color='lightgreen', label='Điểm')
# Thêm nhãn, tiêu đề và chú thích
plt.xticks(x, ten)  # Đặt tên trên trục X
plt.title("Biểu đồ cột: Tuổi và Điểm", fontsize=16)
plt.xlabel("Tên", fontsize=12)
plt.ylabel("Giá trị", fontsize=12)
plt.legend()  # Hiển thị chú thích
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Hiển thị lưới ngang
plt.tight_layout()
# Hiển thị biểu đồ
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu ngẫu nhiên
x = np.random.rand(100)  # 100 giá trị ngẫu nhiên cho trục X
y = np.random.rand(100)  # 100 giá trị ngẫu nhiên cho trục Y

# Vẽ biểu đồ phân tán
plt.scatter(x, y, color='blue', alpha=0.7)  # Màu xanh, độ trong suốt 0.7
plt.title("Biểu đồ phân tán với dữ liệu ngẫu nhiên")
plt.xlabel("Trục X")
plt.ylabel("Trục Y")
plt.grid(True)  # Hiển thị lưới
plt.show()
```
