1. Hãy trình bày dữ liệu biến: GT 
Gợi ý - - 
Lập bảng tần số và tần suất 
Vẽ biểu đồ tần số (cột), biểu đồ tần suất (tròn). 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Tạo dữ liệu mẫu
data = {"GT": ["Nam", "Nữ", "Nam", "Nam", "Nữ", "Nam", "Nữ", "Nam", "Nữ", "Nam"]}
df = pd.DataFrame(data)

# Lập bảng tần số
frequency_table = df["GT"].value_counts().reset_index()
frequency_table.columns = ["Giới tính", "Tần số"]

# Tính tần suất
frequency_table["Tần suất (%)"] = (frequency_table["Tần số"] / len(df)) * 100

# In bảng tần số và tần suất
print(frequency_table)
```

```python
# Vẽ biểu đồ tần số (cột)
plt.figure(figsize=(8, 5))
plt.bar(frequency_table["Giới tính"], frequency_table["Tần số"], color=["blue", "pink"])
plt.title("Biểu đồ Tần số")
plt.xlabel("Giới tính")
plt.ylabel("Tần số")
plt.show()

# Vẽ biểu đồ tần suất (tròn)
plt.figure(figsize=(6, 6))
plt.pie(
    frequency_table["Tần số"],
    labels=frequency_table["Giới tính"],
    autopct="%1.1f%%",
    colors=["skyblue", "lightcoral"],
)
plt.title("Biểu đồ Tần suất")
plt.show()
```

2.Hãy trình bày dữ liệu lần lượt các biến: US_TBM1, US_TBM2 và US_TBM3 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Tạo dữ liệu mẫu
data = {
    "US_TBM1": [8, 7, 6, 8, 7, 6, 8, 7, 6, 6],
    "US_TBM2": [9, 8, 7, 9, 8, 7, 9, 8, 7, 7],
    "US_TBM3": [6, 5, 4, 6, 5, 4, 6, 5, 4, 4],
}
df = pd.DataFrame(data)

# Hàm trình bày dữ liệu
def summarize_and_plot(variable, title):
    # Lập bảng tần số và tần suất
    frequency_table = df[variable].value_counts().reset_index()
    frequency_table.columns = [variable, "Tần số"]
    frequency_table["Tần suất (%)"] = (frequency_table["Tần số"] / len(df)) * 100

    # In bảng
    print(f"*** Thống kê cho {variable} ***")
    print(frequency_table)
    print("\n")

    # Vẽ biểu đồ tần số (cột)
    plt.figure(figsize=(8, 5))
    plt.bar(frequency_table[variable], frequency_table["Tần số"], color="skyblue")
    plt.title(f"Biểu đồ Tần số - {variable}")
    plt.xlabel(variable)
    plt.ylabel("Tần số")
    plt.show()

    # Vẽ biểu đồ tần suất (tròn)
    plt.figure(figsize=(6, 6))
    plt.pie(
        frequency_table["Tần số"],
        labels=frequency_table[variable],
        autopct="%1.1f%%",
        colors=["lightcoral", "skyblue", "lightgreen"],
    )
    plt.title(f"Biểu đồ Tần suất - {variable}")
    plt.show()


# Gọi hàm trình bày cho từng biến
summarize_and_plot("US_TBM1", "US_TBM1")
summarize_and_plot("US_TBM2", "US_TBM2")
summarize_and_plot("US_TBM3", "US_TBM3")
```

3.Hãy trình bày dữ liệu biến DT với các học sinh là nam 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Tạo dữ liệu mẫu
data = {
    "GT": ["Nam", "Nữ", "Nam", "Nữ", "Nam", "Nam", "Nữ", "Nam", "Nam", "Nữ"],
    "DT": ["Kinh", "Kinh", "Tày", "Kinh", "Tày", "Kinh", "Hoa", "Tày", "Kinh", "Kinh"],
}
df = pd.DataFrame(data)

# Lọc dữ liệu cho học sinh là Nam
df_nam = df[df["GT"] == "Nam"]

# Lập bảng tần số và tần suất
frequency_table = df_nam["DT"].value_counts().reset_index()
frequency_table.columns = ["DT", "Tần số"]
frequency_table["Tần suất (%)"] = (frequency_table["Tần số"] / len(df_nam)) * 100

# In bảng tần số và tần suất
print("*** Thống kê cho biến DT với học sinh là Nam ***")
print(frequency_table)
print("\n")

# Vẽ biểu đồ tần số (cột)
plt.figure(figsize=(8, 5))
plt.bar(frequency_table["DT"], frequency_table["Tần số"], color="skyblue")
plt.title("Biểu đồ Tần số - DT (Nam)")
plt.xlabel("Dân tộc (DT)")
plt.ylabel("Tần số")
plt.show()

# Vẽ biểu đồ tần suất (tròn)
plt.figure(figsize=(6, 6))
plt.pie(
    frequency_table["Tần số"],
    labels=frequency_table["DT"],
    autopct="%1.1f%%",
    colors=["lightcoral", "skyblue", "lightgreen", "gold"],
)
plt.title("Biểu đồ Tần suất - DT (Nam)")
plt.show()
```

4.Hãy trình bày dữ liệu biến KV với các học sinh là nam thuộc dân tộc Kinh, có điểm thỏa 
mãn điều kiện (DH1 >= 5.0 và DH2 >= 4.0 và DH3 >= 4.0) 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Tạo dữ liệu mẫu
data = {
    "GT": ["Nam", "Nữ", "Nam", "Nam", "Nữ", "Nam", "Nam", "Nam", "Nữ", "Nam"],
    "DT": ["Kinh", "Kinh", "Tày", "Kinh", "Hoa", "Kinh", "Tày", "Kinh", "Kinh", "Kinh"],
    "KV": ["KV1", "KV2", "KV1", "KV3", "KV1", "KV2", "KV2", "KV3", "KV3", "KV1"],
    "DH1": [6.0, 4.5, 5.0, 5.5, 3.0, 6.0, 4.5, 7.0, 4.0, 6.5],
    "DH2": [5.0, 4.0, 3.5, 5.0, 3.0, 5.5, 4.5, 5.0, 4.0, 6.0],
    "DH3": [4.5, 4.0, 3.5, 5.5, 3.0, 5.0, 4.0, 6.0, 4.5, 5.5],
}

# Tạo DataFrame
df = pd.DataFrame(data)

# Lọc dữ liệu theo các điều kiện
filtered_df = df[
    (df["GT"] == "Nam")
    & (df["DT"] == "Kinh")
    & (df["DH1"] >= 5.0)
    & (df["DH2"] >= 4.0)
    & (df["DH3"] >= 4.0)
]

# Thống kê tần số và tần suất của biến KV
frequency_table = filtered_df["KV"].value_counts().reset_index()
frequency_table.columns = ["KV", "Tần số"]
frequency_table["Tần suất (%)"] = (frequency_table["Tần số"] / len(filtered_df)) * 100

# In bảng thống kê
print("*** Thống kê biến KV với các học sinh thỏa mãn điều kiện ***")
print(frequency_table)
print("\n")

# Vẽ biểu đồ tần số (cột)
plt.figure(figsize=(8, 5))
plt.bar(frequency_table["KV"], frequency_table["Tần số"], color="skyblue")
plt.title("Biểu đồ Tần số - KV (Học sinh thỏa mãn điều kiện)")
plt.xlabel("Khu vực (KV)")
plt.ylabel("Tần số")
plt.show()

# Vẽ biểu đồ tần suất (tròn)
plt.figure(figsize=(6, 6))
plt.pie(
    frequency_table["Tần số"],
    labels=frequency_table["KV"],
    autopct="%1.1f%%",
    colors=["lightcoral", "skyblue", "lightgreen", "gold"],
)
plt.title("Biểu đồ Tần suất - KV (Học sinh thỏa mãn điều kiện)")
plt.show()
```

5.Hãy trình bày dữ liệu lần lượt các biến DH1, DH2, DH3 lớn hơn bằng 5.0 và thuộc khu 
vực 2NT 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Tạo dữ liệu mẫu
data = {
    "KV": ["2NT", "KV1", "2NT", "KV3", "2NT", "2NT", "KV2", "2NT", "KV1", "2NT"],
    "DH1": [6.0, 4.5, 5.0, 5.5, 3.0, 7.0, 4.5, 8.0, 4.0, 6.5],
    "DH2": [5.0, 4.0, 5.5, 6.0, 4.0, 5.5, 4.5, 7.0, 4.5, 6.0],
    "DH3": [5.5, 4.0, 5.0, 6.0, 4.0, 5.0, 4.5, 7.5, 4.5, 5.5],
}

# Tạo DataFrame
df = pd.DataFrame(data)

# Lọc dữ liệu theo điều kiện
filtered_df = df[
    (df["KV"] == "2NT")
    & (df["DH1"] >= 5.0)
    & (df["DH2"] >= 5.0)
    & (df["DH3"] >= 5.0)
]

# Thống kê dữ liệu
statistics = filtered_df[["DH1", "DH2", "DH3"]].describe()

# In bảng thống kê
print("*** Thống kê điểm số thỏa mãn điều kiện ***")
print(statistics)
print("\n")

# Vẽ biểu đồ tần số (cột) cho từng biến
filtered_df[["DH1", "DH2", "DH3"]].plot(kind="bar", figsize=(10, 6))
plt.title("Biểu đồ tần số các biến DH1, DH2, DH3 (KV: 2NT, Điểm >= 5.0)")
plt.xlabel("Học sinh")
plt.ylabel("Điểm số")
plt.legend(["DH1", "DH2", "DH3"])
plt.show()
```
