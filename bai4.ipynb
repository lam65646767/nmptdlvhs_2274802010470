2. Hy to bin phn loi (phanlopt1) cho mn ton (T1) nh sau: 
a. T 0 n di 5 = km (k hiu k)  
b. T 5 n di 7 = trung bnh (k hiu tb) 
c. T 7 n di 8 = kh (k hiu k) 
d. T 8 tr ln = gii (k hiu g) 

```python
import pandas as pd
# D liu mu
T1 = [2.4000, 2.6000, 2.8000, 3.4000, 4.1000, 4.3000, 4.5000, 4.7000, 4.9000,
      5.1000, 5.3000, 5.5000, 5.7000, 5.9000, 6.1000, 7.3000, 8.1000, 9.3000]
# Hm phn loi
def phan_loai_t1(value):
    if value < 5:
        return "k"  # Km
    elif 5 <= value < 7:
        return "tb"  # Trung bnh
    elif 7 <= value < 8:
        return "k"  # Kh
    else:
        return "g"  # Gii
# To danh sch phn loi
phanlopt1 = [phan_loai_t1(x) for x in T1]
# Hin th kt qu di dng DataFrame
df = pd.DataFrame({"T1": T1, "PhanLoaiT1": phanlopt1})
print(df)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# To d liu mu
data = {
    'T1': [2400, 2600, 2800, 3000, 3200, 3400, 3600, 3800, 4000, 4200, 5000, 5200, 6000, 7200, 8000, 9000],
    'Count': [1, 1, 2, 2, 3, 3, 2, 1, 1, 1, 5, 3, 4, 2, 1, 1],
    'PhanlopT1': ['Km', 'Km', 'Trung bnh', 'Trung bnh', 'Kh', 'Kh', 'Kh', 'Trung bnh', 
                  'Trung bnh', 'Km', 'Gii', 'Kh', 'Trung bnh', 'Gii', 'Gii', 'Gii']
}

df = pd.DataFrame(data)

# nh x mu cho cc loi phn lp
colors = {'Gii': 'blue', 'Kh': 'yellow', 'Trung bnh': 'purple', 'Km': 'green'}
df['Color'] = df['PhanlopT1'].map(colors)

# V biu  Drop-line
plt.figure(figsize=(12, 6))
for i in range(len(df)):
    plt.plot([df['T1'][i], df['T1'][i]], [0, df['Count'][i]], color='black', linestyle='-', linewidth=0.8)
    plt.scatter(df['T1'][i], df['Count'][i], color=df['Color'][i], label=df['PhanlopT1'][i] if i == 0 else "")

# Ty chnh trc
plt.xticks(df['T1'], rotation=90)
plt.xlabel("T1")
plt.ylabel("Count")
plt.title("V biu  Drop-line cho bin T1 c phn loi bi bin phanlopT1")
plt.legend(colors.keys(), title="PhanlopT1", loc="upper right")

plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
plt.show()
```
