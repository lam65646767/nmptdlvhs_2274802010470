2. Hãy tạo biến phân loại (phanlopt1) cho môn toán (T1) như sau: 
a. Từ 0 đến dưới 5 = kém (ký hiệu “k”)  
b. Từ 5 đến dưới 7 = trung bình (ký hiệu “tb”) 
c. Từ 7 đến dưới 8 = khá (ký hiệu “k”) 
d. Từ 8 trở lên = giỏi (ký hiệu “g”) 

```python
import pandas as pd
# Dữ liệu mẫu
T1 = [2.4000, 2.6000, 2.8000, 3.4000, 4.1000, 4.3000, 4.5000, 4.7000, 4.9000,
      5.1000, 5.3000, 5.5000, 5.7000, 5.9000, 6.1000, 7.3000, 8.1000, 9.3000]
# Hàm phân loại
def phan_loai_t1(value):
    if value < 5:
        return "k"  # Kém
    elif 5 <= value < 7:
        return "tb"  # Trung bình
    elif 7 <= value < 8:
        return "k"  # Khá
    else:
        return "g"  # Giỏi
# Tạo danh sách phân loại
phanlopt1 = [phan_loai_t1(x) for x in T1]
# Hiển thị kết quả dưới dạng DataFrame
df = pd.DataFrame({"T1": T1, "PhanLoaiT1": phanlopt1})
print(df)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Tạo dữ liệu mẫu
data = {
    'T1': [2400, 2600, 2800, 3000, 3200, 3400, 3600, 3800, 4000, 4200, 5000, 5200, 6000, 7200, 8000, 9000],
    'Count': [1, 1, 2, 2, 3, 3, 2, 1, 1, 1, 5, 3, 4, 2, 1, 1],
    'PhanlopT1': ['Kém', 'Kém', 'Trung bình', 'Trung bình', 'Khá', 'Khá', 'Khá', 'Trung bình',
                  'Trung bình', 'Kém', 'Giỏi', 'Khá', 'Trung bình', 'Giỏi', 'Giỏi', 'Giỏi']
}

df = pd.DataFrame(data)

# Nhóm dữ liệu theo 'PhanlopT1'
grouped = df.groupby('PhanlopT1')

# Ánh xạ màu cho các loại phân lớp
colors = {'Giỏi': 'blue', 'Khá': 'yellow', 'Trung bình': 'purple', 'Kém': 'green'}

# Vẽ biểu đồ đường Multiple Line
plt.figure(figsize=(12, 6))
for name, group in grouped:
    # Sắp xếp nhóm theo T1 để đảm bảo đường nối đúng thứ tự
    group = group.sort_values('T1')
    plt.plot(group['T1'], group['Count'], label=name, color=colors[name], marker='o')

# Tùy chỉnh trục và nhãn
plt.xticks(df['T1'], rotation=90)
plt.yticks(range(0, 6))  # Hiển thị các giá trị 0, 1, 2, 3, 4, 5 trên trục y
plt.xlabel("T1")
plt.ylabel("Count")
plt.title("Vẽ biểu đồ đường Multiple Line cho biến T1 được phân loại bởi biến phanlopT1")
plt.legend(title="PhanlopT1", loc="upper right")

plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Tạo dữ liệu mẫu
data = {
    'T1': [2400, 2600, 2800, 3000, 3200, 3400, 3600, 3800, 4000, 4200, 5000, 5200, 6000, 7200, 8000, 9000],
    'Count': [1, 1, 2, 2, 3, 3, 2, 1, 1, 1, 5, 3, 4, 2, 1, 1],
    'PhanlopT1': ['Kém', 'Kém', 'Trung bình', 'Trung bình', 'Khá', 'Khá', 'Khá', 'Trung bình', 
                  'Trung bình', 'Kém', 'Giỏi', 'Khá', 'Trung bình', 'Giỏi', 'Giỏi', 'Giỏi']
}

df = pd.DataFrame(data)

# Ánh xạ màu cho các loại phân lớp
colors = {'Giỏi': 'blue', 'Khá': 'yellow', 'Trung bình': 'purple', 'Kém': 'green'}
df['Color'] = df['PhanlopT1'].map(colors)

# Vẽ biểu đồ Drop-line
plt.figure(figsize=(12, 6))
for i in range(len(df)):
    plt.plot([df['T1'][i], df['T1'][i]], [0, df['Count'][i]], color='black', linestyle='-', linewidth=0.8)
    plt.scatter(df['T1'][i], df['Count'][i], color=df['Color'][i], label=df['PhanlopT1'][i] if i == 0 else "")

# Tùy chỉnh trục
plt.xticks(df['T1'], rotation=90)
plt.xlabel("T1")
plt.ylabel("Count")
plt.title("Vẽ biểu đồ Drop-line cho biến T1 được phân loại bởi biến phanlopT1")
plt.legend(colors.keys(), title="PhanlopT1", loc="upper right")

plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
plt.show()
```
